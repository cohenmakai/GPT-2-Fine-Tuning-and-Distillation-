## First train dreambooth on thumbs up images

```python
!export MODEL_NAME="stabilityai/stable-diffusion-2"
!export INSTANCE_DIR="thumbs_up"
!export OUTPUT_DIR="db_thumbsup"

!accelerate launch --mixed_precision="fp16" train_dreambooth.py \
  --pretrained_model_name_or_path=$MODEL_NAME  \
  --instance_data_dir=$INSTANCE_DIR \
  --output_dir=$OUTPUT_DIR \
  --instance_prompt="a photo of person doing thumbs up" \
  --resolution=768 \
  --train_batch_size=1 \
--gradient_accumulation_steps=2 --gradient_checkpointing \
  --use_8bit_adam \
  --checkpointing_steps=100 \
  --learning_rate=1e-4 \
  --report_to="wandb" \
  --lr_scheduler="cosine_with_restarts" \
  --lr_warmup_steps=0 \
  --max_train_steps=3000 \
  --validation_prompt="An old lady doing thumbs up" \
  --validation_steps=50 \
  --seed="0" \
  --push_to_hub --hub_token="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"
```

## Then fine tune on Abhi images using LoRA

```python
!export MODEL_NAME="db_thumbsup"
!export INSTANCE_DIR="abhi"
!export OUTPUT_DIR="db_abhi_tu"

!accelerate launch train_dreambooth_lora.py \
  --pretrained_model_name_or_path=$MODEL_NAME  \
  --instance_data_dir=$INSTANCE_DIR \
  --output_dir=$OUTPUT_DIR \
  --instance_prompt="a photo of Abhi" \
  --resolution=768 \
  --train_batch_size=1 \
  --gradient_accumulation_steps=1 \
  --checkpointing_steps=100 \
  --learning_rate=1e-4 \
  --report_to="wandb" \
  --lr_scheduler="cosine_with_restarts" \
  --lr_warmup_steps=0 \
  --max_train_steps=1000 \
  --validation_prompt="A photo of Abhi doing thumbs up" \
  --validation_epochs=20 \
  --seed="0" \
  --push_to_hub --hub_token="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"
```
